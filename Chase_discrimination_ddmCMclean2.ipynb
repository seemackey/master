## Fitting go/no-go using the chi-qsuare approach ##

http://ski.clps.brown.edu/hddm_docs/demo_gonogo.html

 demo of: https://www.biorxiv.org/content/10.1101/447656v1.full
 
 published paper: https://elifesciences.org/articles/54014

Columns are 

1) noise modulation frequency 1, 

2) noise mod freq 2, 

3) lever release (1 go, 2 no-go), (note: changed no go to 0 in code)

4) correct or incorrect (0 is incorrect),\ 

5) reaction time from 0-1 Sec. 999 indicates a no-go response.

added: 6) condition (1=go is correct resp, 0=no go is correct resp)

Some notes:

No-gos are all 999 ms. Noise mod freq 2 is the comparison stimulus, and sufficient for knowing the trial type (I just left noise 1 in there just in case). No-gos are always incorrect if noise 1 and noise 2 are different. No-gos are correct if noise 1 and 2 are same (catch trials). I think that should be all you need to know.

# FUNCTIONS FOR FITTING CHI-SQUARE TO DATA #

```python
quantiles=[0, 0.1, 0.3, 0.5, 0.7, 0.9]
```

```python
def fit_subject(dataframe,quants):

## Simulates stim-coded data
    model = hddm.HDDMStimCoding(dataframe,stim_col='stim_col',split_param='v', drift_criterion=True, bias=True, p_outlier=0, include=('z','a','v'),depends_on={'v':['NMF2']})
    model.optimize('gsquare', quantiles=quants, n_runs=10)
    

    return model 
```

```python
def get_model_residuals(dataframe,quants):
    model = hddm.HDDMStimCoding(dataframe,stim_col='stim',split_param='v', drift_criterion=False, bias=True, p_outlier=0, include=('z','a','v','t'),depends_on={'v':['NMF2'],'t':['if_catch_trial'],'a':['if_catch_trial']})
    model.optimize('gsquare', quants, n_runs=10)
    res = pd.concat((pd.DataFrame([model.values], index=[dataframe.subj_idx.unique()[0]]), pd.DataFrame([model.bic_info], index=[dataframe.subj_idx.unique()[0]])), axis=1)

    return res
```

```python
# set no-go responses to have NaN RTs, necessary for go/nogo tasks
go_nogo = True 
df=df.copy(deep=True)

if go_nogo:
    df.loc[df["response"]==0, 'rt'] = np.NaN
    
```

# FIT CHI-SQUARE MODELS TO DATA #

```python
# fit chi-square:
quantiles = [0,.1, .3, .5, .7, .9]
params_fitted = pd.concat(Parallel(n_jobs=n_subjects)(delayed(get_model_residuals)(data[1], quantiles) # get the model out too
                                                                                  for data in df.groupby('subj_idx')))
#params_fitted.drop(['bic', 'likelihood', 'penalty', 'z_trans(0)', 'z_trans(1)'], axis=1, inplace=True) 
print(params_fitted.head())
```

# DATA SIMULATION #

```python

##function for generating synthetic data## this is for the go/no-go model where v depends on NMF2
def gen_simdata_discrimination_DDM2(params_fitted,df):
    subject_list=list(params_fitted.index.unique())
    n_subjects=len(subject_list)
    params_all=[]
    #NMF2s=[20,22,24,28,36,52,84] # remove 21 hz?
    NMF2s=[20,24,28,36,44,52,84] # remove 21 hz?
    n_NMF2s=len(NMF2s)
    n_NMF2vals_subj=[]
    params_condition_list=[]
    params_NMF2vals=[]
    
    for i_subject in range(n_subjects):
        sub_params=params_fitted.loc[subject_list[i_subject]]
        for i_NMF2 in range(n_NMF2s):
            n_sub_NMF2=len(df[(df.subj_idx==subject_list[i_subject])&(df.NMF2==NMF2s[i_NMF2])])
            if n_sub_NMF2>0:
                if NMF2s[i_NMF2]!=min(NMF2s):
                    params_all.append({'v':sub_params.loc['v({}.0)'.format(NMF2s[i_NMF2])], 'a':sub_params.loc['a(0.0)'], 't':sub_params.loc['t(0.0)'],'z':sub_params.loc['z']})
                else:
                    params_all.append({'v':sub_params.loc['v({}.0)'.format(NMF2s[i_NMF2])], 'a':sub_params.loc['a(1.0)'], 't':sub_params.loc['t(1.0)'],'z':sub_params.loc['z']})
                params_condition_list.append('v({}.0)'.format(NMF2s[i_NMF2]))
                #params_condition_list.append('v({}.0)'.format(NMF2s[i_NMF2]))
                params_NMF2vals.append([NMF2s[i_NMF2],subject_list[i_subject]])
                #params_NMF2vals.append([NMF2s[i_NMF2],subject_list[i_subject]])
                n_NMF2vals_subj.append(n_sub_NMF2)
                #n_NMF2vals_subj.append(n_sub_NMF2/2)
                #n_NMF2vals_subj.append(n_sub_NMF2/2)    
                
    
    df_simdata=pd.DataFrame(data=None,columns=['rt','response','subj_idx','NMF2'])


    for i_conditions in range(len(params_all)):
        simdata,simdata_params=hddm.generate.gen_rand_data(params=params_all[i_conditions],size=n_NMF2vals_subj[i_conditions], subjs=1)
        simdata.rename(columns = {"condition":"NMF2"}, inplace = True)
        condcol=pd.Series(data=[params_NMF2vals[i_conditions][0]]*len(simdata))
        simdata['NMF2']=condcol
        subnum=pd.Series(data=[params_NMF2vals[i_conditions][1]]*len(simdata))
        simdata['subj_idx']=subnum
        colsa = list(set(df_simdata.columns) & set(simdata.columns)) #get data from columns that sim data and df have in common
        df_simdata=df_simdata.append(simdata[colsa], ignore_index=True,sort=False) #add sim data to relevant columns 

    return df_simdata
```

```python
# run simulate data fxn for regular model
simdata = gen_simdata_discrimination_DDM2(params_fitted,df)
```

```python
# export sim data from regular model
simdata.to_csv('simdata_chisquare_50ms_withDarioV2.csv')
```
